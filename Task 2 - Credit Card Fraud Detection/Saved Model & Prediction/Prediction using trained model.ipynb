```python
import pickle
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',200)
from sklearn.metrics import classification_report,accuracy_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
with open('rf_model','rb') as file:
    rf = pickle.load(file)
```

```python
test_data = pd.read_csv('Dataset/fraudTest.csv')
test_data.isnull().sum()
```

```python
test_data.drop_duplicates(inplace = True)
```

```python
test_data.shape
```

```python
new_test_data = test_data[['gender','category','amt', 'lat', 'long','city_pop', 'unix_time', 'merch_lat', 'merch_long','is_fraud']]
```

```python
new_test_data.head()
```

### Categorical

```python
# One Hot Encoding

categ_cols = ['gender','category']

ohe = OneHotEncoder()

test_data_encoded = ohe.fit_transform(test_data[categ_cols]).toarray()
```

```python
feature_labels = ohe.categories_
feature_labels = np.concatenate(feature_labels)
```

### Feature Scaling

```python
numerical_cols = ['amt', 'lat', 'long','city_pop', 'unix_time', 'merch_lat', 'merch_long']

scaler = StandardScaler()

test_scaled = scaler.fit_transform(new_test_data[numerical_cols])
```

```python
test_data_final = pd.concat([pd.DataFrame(test_data_encoded, columns=feature_labels),pd.DataFrame(test_scaled, columns=numerical_cols)],axis=1).drop(['F','travel'],axis=1)
test_data_final.head()
```

```python
new_test_data.head()
```

```python
new_test_data['Predicted'] = rf.predict(test_data_final)
new_test_data.head()
```

```python
print('Classification Report: \n', classification_report(new_test_data['is_fraud'], new_test_data['Predicted']))
print('Accuracy Score: ', accuracy_score(new_test_data['is_fraud'], new_test_data['Predicted']))
```

```python
new_test_data.to_csv('Predicted.csv',index=False)
```
